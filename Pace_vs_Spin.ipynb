```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import date
pd.set_option("display.max_rows", None, "display.max_columns", None)
pd.set_option('display.width', 1000)
player_name = 'Glenn Maxwell'
player_role = 'batsman' # 'batsman' or 'bowler' or 'all'
format = 'TWENTY20' # 'Test' or 'ODI' or 'T20I' or 'TWENTY20' or 'T10' or 'all'
grouping = 'opp_player_type' #overs or opp_player_type or opp_player opp_team or inning or year or venue or batpos
start_date = '2017-01-01'
end_date = date.today()
start_over = 0
end_over = 21
headings = []
column1 = []
column2 = []
column3 = []
column4 = []
column5 = []
column6 = []
column7 = []
column8 = []
column9 = []
column10 = []
column11 = []
URL = (f'http://www.cricmetric.com/playerstats.py?player={player_name}&role={player_role}&format={format}&groupby={grouping}&start_date={start_date}&end_date={end_date}&start_over={start_over}&end_over={end_over}')
#URL = 'http://www.cricmetric.com/playerstats.py?player=B+Kumar&role=bowler&format=TWENTY20&groupby=opp_player_type&start_date=2018-01-01&end_date=2021-05-15&start_over=0&end_over=9999'
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'lxml')
info = soup.find('table')
#print(info)
column = info.find_all('tbody')
#print(column)
info1 = info.find_all('tr')
info2 = info.find_all('th')
info3 = info.find_all('td')
k = len(info2)
for i in range (0,len(info2)):
    #print(info2[i].text)
    headings.append(info2[i].text)
for i in range (0,len(info2)):
    #print(info3[i].text)
    column1.append(info3[i].text)
for i in range (k,2*k):
    #print(info3[i].text)
    column2.append(info3[i].text)
for i in range (2*k,3*k):
    #print(info3[i].text)
    column3.append(info3[i].text)
final_df = pd.DataFrame(columns = ['Versus Player Type', 'Innings', 'Runs', 'Balls', 'Outs', 'Avg', 'SR', 'HS', '50', '100', '4s', '6s', 'Dot%'])
#final_df = pd.DataFrame(column1).T
#final_df.columns = headings
final_df.loc[len(final_df.index)] = column1
final_df.loc[len(final_df.index)] = column2
final_df.loc[len(final_df.index)] = column3
if len(info3) > 3*k:
    for i in range (3*k,4*k):
        column4.append(info3[i].text)
    final_df.loc[len(final_df.index)] = column4
if len(info3) > 4*k:
    for i in range (4*k,5*k):
        column5.append(info3[i].text)
    final_df.loc[len(final_df.index)] = column5
if len(info3) > 5*k:
    for i in range (5*k,6*k):
        column6.append(info3[i].text)
    final_df.loc[len(final_df.index)] = column6
if len(info3) > 6*k:
    for i in range (6*k,7*k):
        column7.append(info3[i].text)
    final_df.loc[len(final_df.index)] = column7
if len(info3) > 7*k:
    for i in range (7*k,8*k):
        column8.append(info3[i].text)
    final_df.loc[len(final_df.index)] = column8
if len(info3) > 8*k:
    for i in range (8*k,9*k):
        column9.append(info3[i].text)
    final_df.loc[len(final_df.index)] = column9
if len(info3) > 9*k:
    for i in range (9*k,10*k):
        column10.append(info3[i].text)
    final_df.loc[len(final_df.index)] = column10
if len(info3) > 10*k:
    for i in range (10*k,11*k):
        column11.append(info3[i].text)
    final_df.loc[len(final_df.index)] = column11
final_df.rename(columns = final_df.iloc[0])
print(player_role.capitalize(), "Name:",player_name)
final_df.drop(final_df.tail(1).index,inplace = True)
runs = final_df['Runs']
for i in range(0,len(runs)):
    num = runs[i]
    if(num[2] == ","):
        runs[i] = num[0:2]+num[3:6]
final_df['Runs'] = runs
balls = final_df['Balls']
for i in range(0,len(balls)):
    num = balls[i]
    if(num[2] == ","):
        balls[i] = num[0:2]+num[3:6]
final_df['Balls'] = balls
final_df['Innings'] = final_df['Innings'].astype(int)
final_df['Runs'] = final_df['Runs'].astype(int)
final_df['Balls'] = final_df['Balls'].astype(int)
final_df['Outs'] = final_df['Outs'].astype(int)
final_df['4s'] = final_df['4s'].astype(int)
final_df['6s'] = final_df['6s'].astype(int)
final_df['Dot%'] = final_df['Dot%'].astype(float)
final_df['Dots'] = round((final_df['Balls']*final_df['Dot%'])/100,0)
#print(final_df)
spin_df = final_df[(final_df['Versus Player Type']=='Right-arm Offbreak ') | (final_df['Versus Player Type'] == 'Right-arm Legbreak ') | (final_df['Versus Player Type']=='Left-arm Orthodox ') | (final_df['Versus Player Type']=='Left-arm Chinaman ')]
pace_df = final_df[(final_df['Versus Player Type']=='Right-arm Fast ') | (final_df['Versus Player Type'] == 'Right-arm Medium ') | (final_df['Versus Player Type']=='Left-arm Fast ') | (final_df['Versus Player Type']=='Left-arm Medium ')]
#print(spin_df)
#print(pace_df)
pace_innings = pace_df['Innings'].sum()
pace_runs = pace_df['Runs'].sum()
pace_balls = pace_df['Balls'].sum()
pace_outs = pace_df['Outs'].sum()
pace_4s = pace_df['4s'].sum()
pace_6s = pace_df['6s'].sum()
pace_dots = pace_df['Dots'].sum()
pace_avg = round(pace_runs / pace_outs,2)
pace_sr = round((pace_runs/pace_balls)*100,2)
pace_bpb = round(pace_balls/(pace_4s + pace_6s),2)
pace_dbp = round((pace_dots/pace_balls)*100,2)
spin_innings = spin_df['Innings'].sum()
spin_runs = spin_df['Runs'].sum()
spin_balls = spin_df['Balls'].sum()
spin_outs = spin_df['Outs'].sum()
spin_4s = spin_df['4s'].sum()
spin_6s = spin_df['6s'].sum()
spin_dots = spin_df['Dots'].sum()
spin_avg = round(spin_runs / spin_outs,2)
spin_sr = round((spin_runs/spin_balls)*100,2)
spin_bpb = round(spin_balls/(spin_4s + spin_6s),2)
spin_dbp = round((spin_dots/spin_balls)*100,2)
new_df = pd.DataFrame({'Bowling Type':['Pace', 'Spin'], 'Innings':[pace_innings, spin_innings], 'Runs':[pace_runs, spin_runs], 'Balls':[pace_balls, spin_balls], 'Average':[pace_avg, spin_avg], 'SR':[pace_sr, spin_sr], 'BPB':[pace_bpb, spin_bpb], 'DBP':[pace_dbp, spin_dbp]})
print(new_df)

```
